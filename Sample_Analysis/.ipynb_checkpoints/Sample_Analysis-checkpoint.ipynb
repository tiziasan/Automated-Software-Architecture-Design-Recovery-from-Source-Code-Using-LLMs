```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel file and the AN1 sheet data
file_path = 'Sample_Analysis.xlsx'  # Replace with your file path
excel_data = pd.ExcelFile(file_path)
an1_data = excel_data.parse('AN1')

# Create separate box plots for LOC and Nr.Classes using light blue color with thicker median lines
plt.figure(figsize=(10, 6))

# Define a light blue color for both box plots
box_color = 'lightblue'

# Plot for LOC with light blue color and thick median line
plt.subplot(1, 2, 1)
box_loc = plt.boxplot(an1_data['LOC'], patch_artist=True)
#plt.title('Box Plot for LOC')
plt.ylabel('Values')
plt.xticks([1], ['LOC'])

for patch in box_loc['boxes']:
    patch.set_facecolor(box_color)
for median in box_loc['medians']:
    median.set_linewidth(2.5)  # Set median line thickness

# Plot for Nr.Classes with light blue color and thick median line
plt.subplot(1, 2, 2)
box_classes = plt.boxplot(an1_data['Nr.Classes'], patch_artist=True)
#plt.title('Box Plot for Nr.Classes')
#plt.ylabel('Values')
plt.xticks([1], ['#Classes'])
for patch in box_classes['boxes']:
    patch.set_facecolor(box_color)
for median in box_classes['medians']:
    median.set_linewidth(2.5)  # Set median line thickness

plt.tight_layout()

# Save the figure as a PNG file
plt.savefig('Sample_1_box_plots_LOC_Classes.png', format='png', dpi=300)

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel file and the Full sheet data
file_path = 'Sample_Analysis.xlsx'  # Replace with your actual file path
excel_data = pd.ExcelFile(file_path)
full_data = excel_data.parse('Full')

# Define pastel colors for specific samples
pastel_colors = ['lightcoral' if 'AN1' in sample else 'lightgreen' if 'AN2' in sample else 'lightblue' if 'AN3' in sample else 'lightgrey'
                 for sample in full_data['Sample']]

# Create a bar chart for the LOC column in the Full sheet with pastel colors and no title
plt.figure(figsize=(12, 6))

# Plot the bar chart for LOC with specific pastel colors for Sample1, Sample2, Sample3
plt.bar(full_data['Sample'], full_data['LOC'], color=pastel_colors, edgecolor='black')
plt.xlabel('Sample')
plt.ylabel('LOC')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.grid(False)  # Remove grid

# Create custom legend with pastel colors
legend_labels = ['Analysis 1', 'Analysis 2', 'Analysis 3']
legend_colors = ['lightcoral', 'lightgreen', 'lightblue']
custom_legend = [plt.Line2D([0], [0], color=color, lw=4) for color in legend_colors]
plt.legend(custom_legend, legend_labels, title='Legend')

plt.tight_layout()

# Save the figure as a PNG file
plt.savefig('LOC_bar_chart_full_sheet.png', format='png', dpi=300)

# Display the plot
plt.show()
```
